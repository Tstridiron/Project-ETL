```python
import requests
import json
from pprint import pprint
import pandas as pd
from config import password
from config_db import password_db
from sqlalchemy import create_engine
```

```python
# import requests from FIFA API to get the World Cup teams information

url = "https://montanaflynn-fifa-world-cup.p.rapidapi.com/teams"

headers = {
    'x-rapidapi-host': "montanaflynn-fifa-world-cup.p.rapidapi.com",
    'x-rapidapi-key': password,
    'accepts': "json"
    }

response = requests.request("GET", url, headers=headers, verify=False)
data = response.json()

print(json.dumps(data, indent=4, sort_keys=True))
```

```python
# Check how much data contains in worldcup teams API

teams_code = data[0]["title"]
print(teams_code)

teamscount = len(data)
teamscount
```

```python
# Append the required information in lists

code = []
title = []
key = []


id = 0
for each in range(teamscount):
    data = response.json()
    code.append(data[id]['code'])
    title.append(data[id]['title'])
    key.append(data[id]['key'])
    
    id +=1
    
print(f"{code}{title}{key}")
    
    
```

```python
# get data into dataframe

team_df = pd.DataFrame({
    "code":code,
    "title":title,
    "key":key,
                         })

team_df
```

```python
# rename the code field into country so it can join with other datasets

team_df_renamed = team_df

team_df_renamed1 = team_df_renamed.rename(columns={'code':'Country'})

team_df_renamed1
```

```python
# Get players data from csv file extracted sqlite manually

players_data = pd.read_csv("Resources/Players.csv")
players_data
```

```python
# transform players data by dropping unecessary columns

players_data = players_data.drop(['id','player_fifa_api_id'],axis = 1)
players_data
```

```python
# Set index for players dataset

players_data_new1 = players_data.set_index('player_api_id')
players_data_new1
```

```python
# pull players attributes dataset adquired from csv

playerattr_df = pd.read_csv("Resources/Players_Attributes.csv")
playerattr_df
```

```python
# transform player attribute dataset

playerattr_df.dropna(axis=0, how='any', inplace=True)
playerattr_df.drop(["id","player_fifa_api_id","attacking_work_rate", "defensive_work_rate","short_passing","dribbling","long_passing",
                    "ball_control","acceleration","sprint_speed","agility","reactions","balance","jumping","stamina",
                    "strength","aggression","interceptions","positioning","vision","penalties","marking","standing_tackle","sliding_tackle"],
                   axis = 1, inplace = True)
playerattr_df.head()
```

```python
# continue with ETL work for player attribute

playerattr_df = playerattr_df.set_index('player_api_id')
playerattr_df
```

```python
# get the latest date for each player id, so it can be uniquely identified (only interested on most recent information)

playerattr_df_new = playerattr_df.sort_values('date').groupby('player_api_id').max()
playerattr_df_new
```

```python
# extract data from penalty dataset (this is the key information for our use case. Intention is to evaluate Goalkeepers
# in corresponding teams with their performance)

keeper_df = pd.read_csv("Resources/GKPenalty.csv")
keeper_df
```

```python
# ETL on keepers data by trimming unecessary data off

keeper_df.dropna(axis=0, how='any', inplace=True)
keeper_df.drop(["Current team", "Born"], axis = 1, inplace = True)
keeper_df
```

```python
# More transformations in preparation for merging datasets

keeper_df_rename = keeper_df.rename(columns={'Goalkeeper':'player_name'})
keeper_df_rename
```

```python
# Transformation continues to lower case the names to get higher match between datasets

keeper_df_rename['player_name'] = keeper_df_rename['player_name'].str.lower()
keeper_df_rename
```

```python
# merging datasets of players with players attribute along with transformations

merged_players = pd.merge(players_data_new1, playerattr_df_new, how = 'left', on='player_api_id')
merged_players.dropna(axis=0, how='any',inplace=True)
merged_players['player_name'] = merged_players['player_name'].str.lower()
merged_players
```

```python
# merging prior merging results to keepers information

merged_keepers = pd.merge(keeper_df_rename, merged_players, how = 'left', on='player_name')
merged_keepers
```

```python
# merging data with FIFA world cup content to identify the national team that each keeper played

final_merged_keepers = pd.merge(merged_keepers, team_df_renamed1, how = 'left', on='Country')
final_merged_keepers
```

```python
# further transformations to drop unecessary columns

final_merged_keepers.drop(["key","birthday"], axis = 1, inplace = True)
final_merged_keepers
```

```python
# Preparing data to align same structure as in database

final_keepers_renamed = final_merged_keepers.rename(columns={'Country':'country','NotConceded':'not_conceded',\
                                                             'Conceded':'conceded',\
                                                             'PKsFaced':'pksfaced', \
                                                             'Saves':'saves', \
                                                            'gk_diving':'diving','gk_handling': 'handling',\
                                                             'gk_kicking':'kicking','gk_positioning':'positioning', \
                                                            'gk_reflexes':'reflexes'})

final_keepers_renamed
```

```python
# creating connection to postgress sql database

connection_string = 'postgres:'+password_db+'@localhost:5432/keepers_db'

engine=create_engine(f"postgresql://{connection_string}")
```

```python
# verify table if exists
engine.table_names()
```

```python
# push data to table in database
final_keepers_renamed.to_sql(name='keepers_records', con=engine, if_exists='append', index=True)
```
